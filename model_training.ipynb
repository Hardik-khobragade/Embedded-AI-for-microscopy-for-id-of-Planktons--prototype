```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.yaml")  # build a new model from YAML
model = YOLO("yolo11n.pt")  # load a pretrained model (recommended for training)
model = YOLO("yolo11n.yaml").load("yolo11n.pt")  # build from YAML and transfer weights

# Train the model
results = model.train(data="file.yaml", epochs=20, imgsz=640)
```

```python
from ultralytics import YOLO
import cv2

# Load trained model
model = YOLO("C:/Users/ASUS/runs/detect/train/weights/best.pt")  # update path

# Load video
cap = cv2.VideoCapture("videoplayback (2).mp4")  # or 0 for webcam

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert to grayscale (black and white)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Convert back to 3-channel for YOLO (expects 3-channel input)
    gray_3ch = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

    # Run detection
    results = model(gray_3ch)

    # Plot results on frame
    annotated_frame = results[0].plot()

    # Show video
    cv2.imshow("YOLOv11 Detection - B/W", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()

```
